```python
import pandas as pd
import json
import pickle
import yaml
import os

# ✅ Correct paths
input_path = r"C:\Users\NXTWAVE\Downloads\Smart Market Intelligence & Business Insights System\megambmarketname_n.xls"

output_dir = r"C:\Users\NXTWAVE\Downloads\Smart Market Intelligence & Business Insights System"

# Create folder if not exists
os.makedirs(output_dir, exist_ok=True)

# -------------------------------
# 1. Load Excel File
# -------------------------------
df = pd.read_excel(input_path)

print("Dataset Loaded Successfully!")
print(df.head())

# -------------------------------
# 2. Save as HDF5 (.h5)
# -------------------------------
h5_path = os.path.join(output_dir, "megambmarketname.h5")
df.to_hdf(h5_path, key='df', mode='w')

# -------------------------------
# 3. Save as Pickle (.pkl)
# -------------------------------
pkl_path = os.path.join(output_dir, "megambmarketname.pkl")
df.to_pickle(pkl_path)

# -------------------------------
# 4. Save as JSON (.json)
# -------------------------------
json_path = os.path.join(output_dir, "megambmarketname.json")
df.to_json(json_path, orient="records", indent=4)

# -------------------------------
# 5. Save as YAML (.yaml)
# -------------------------------
yaml_path = os.path.join(output_dir, "megambmarketname.yaml")

with open(yaml_path, "w") as file:
    yaml.dump(df.to_dict(orient="records"), file)

# -------------------------------
# Done
# -------------------------------
print("\nAll files saved successfully!")
print(f"H5: {h5_path}")
print(f"PKL: {pkl_path}")
print(f"JSON: {json_path}")
print(f"YAML: {yaml_path}")
```
